1. **Асинхронный таймер:**
   - Напишите программу с использованием `asyncio`, которая в течение определенного времени (например, 10 секунд) выводит сообщение каждую секунду. Используйте `asyncio.sleep` для создания таймера.

```python
import asyncio

async def async_timer(duration: int = 10) -> None:
    """Выводит уведомление каждую секунду в течение указанного времени."""
    for second in range(1, duration + 1):
        await asyncio.sleep(1)
        print(f"Прошло {second} секунд")

async def main() -> None:
    print("Асинхронный таймер запущен")
    await async_timer()
    print("Таймер завершён")

await main()
```
